```python
# 05-1
# calculator.py (전역 변수 사용 예)
result = 0

def add(num):
    global result
    result += num
    return result

print(add(3))
print(add(4))


# calculator2.py (여러 계산기 함수)
result1 = 0
result2 = 0

def add1(num):
    global result1
    result1 += num
    return result1

def add2(num):
    global result2
    result2 += num
    return result2

print(add1(3))
print(add1(4))
print(add2(3))
print(add2(7))


# calculator3.py (클래스 사용)
class Calculator:
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result += num
        return self.result

    def sub(self, num):
        self.result -= num
        return self.result


cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(3))
print(cal2.add(7))


# FourCal 클래스
class FourCal:
    def __init__(self, first, second):
        self.first = first
        self.second = second

    def setdata(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        return self.first + self.second

    def mul(self):
        return self.first * self.second

    def sub(self):
        return self.first - self.second

    def div(self):
        return self.first / self.second


a = FourCal(4, 2)
b = FourCal(3, 8)

print(a.add())
print(a.mul())
print(a.sub())
print(a.div())

print(b.add())
print(b.mul())
print(b.sub())
print(b.div())


# 상속
class MoreFourCal(FourCal):
    def pow(self):
        return self.first ** self.second


c = MoreFourCal(4, 2)
print(c.add())
print(c.mul())
print(c.sub())
print(c.div())
print(c.pow())


# 나눗셈 안전 처리
class SafeFourCal(FourCal):
    def div(self):
        if self.second == 0:
            return 0
        return self.first / self.second


d = SafeFourCal(4, 0)
print(d.div())


# 클래스 변수 예제
class Family:
    lastname = "김"


print(Family.lastname)

x = Family()
y = Family()

print(x.lastname)
print(y.lastname)

Family.lastname = "박"
print(x.lastname)
print(y.lastname)

x.lastname = "최"
print(x.lastname)
print(Family.lastname)
print(y.lastname)
```

```python
# 05-4 
PATH = "./09-python.txt"

# 1. 기본 예외 처리
try:
    f = open("나없는파일", "r")
except FileNotFoundError as e:
    print("파일 오류:", e)

try:
    result = 4 / 0
except ZeroDivisionError as e:
    print("수학 오류:", e)

try:
    a = [1, 2, 3]
    print(a[3])
except IndexError as e:
    print("인덱스 오류:", e)

print("-" * 30)

# 2. 여러 예외 처리
try:
    a = [1, 2]
    print(a[3])
    4 / 0
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
except IndexError:
    print("인덱싱 할 수 없습니다.")

print("-" * 30)

# 3. 여러 예외를 하나로 처리
try:
    a = [1, 2]
    print(a[3])
    4 / 0
except (ZeroDivisionError, IndexError) as e:
    print("에러 발생:", e)

print("-" * 30)

# 4. try ~ else
try:
    age = int("20")
except ValueError:
    print("입력이 정확하지 않습니다.")
else:
    if age <= 18:
        print("미성년자는 출입금지입니다.")
    else:
        print("환영합니다.")

print("-" * 30)

# 5. try ~ finally
try:
    f = open(PATH, "w")
    f.write("Hello Python")
finally:
    f.close()
    print("파일이 정상적으로 닫혔습니다.")

print("-" * 30)

# 6. 파일 처리 + continue
students = ["김철수", "이영희", "박민수", "최유진"]

for student in students:
    try:
        with open(f"{student}_성적.txt", "r") as f:
            score = f.read()
            print(f"{student}의 성적: {score}")
    except FileNotFoundError:
        print(f"{student}의 성적 파일이 없습니다. 건너뜁니다.")
        continue

print("-" * 30)

# 7. pass 사용
try:
    f = open("설정파일.txt", "r")
    config = f.read()
    f.close()
except FileNotFoundError:
    pass

print("프로그램이 정상적으로 실행됩니다.")

print("-" * 30)

# 8. raise 사용 (추상 메서드)
class Bird:
    def fly(self):
        raise NotImplementedError

class Eagle(Bird):
    def fly(self):
        print("very fast")

eagle = Eagle()
eagle.fly()

print("-" * 30)

# 9. 사용자 정의 예외
class MyError(Exception):
    def __str__(self):
        return "허용되지 않는 별명입니다."

def say_nick(nick):
    if nick == "바보":
        raise MyError()
    print(nick)

try:
    say_nick("천사")
    say_nick("바보")
except MyError as e:
    print(e)
```
