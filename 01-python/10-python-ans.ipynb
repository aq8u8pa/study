## 1. 클래스
### 설명
* 아래 코드는 전역 변수를 사용한 계산기입니다.
* 전역 변수 대신 클래스와 인스턴스 변수를 사용하도록 수정하세요.
    * 클래스 정의
    * `self.result` 사용

### 예시 출력
```sh
3  
7
```

````python
class Calculator:
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result += num
        return self.result

cal = Calculator()
print(cal.add(3))
print(cal.add(4))
````


## 2. 인스턴스

### 설명
* 같은 클래스로 만든 객체라도 각자의 상태는 서로 영향을 주면 안 됩니다.
    * 인스턴스 변수 개념 이해
* 아래 코드는 의도적으로 잘못 작성된 계산기 클래스입니다. 두 개의 계산기를 만들어도 결과가 서로 섞이는 문제가 있습니다.
* 이 코드를 수정해서 각 계산기 인스턴스가 자신의 결과만 관리하도록 만드세요.

### 예시 출력
* cal1과 cal2는 서로 영향을 주지 않아야 합니다.
```sh
3
7
3
10
```

````python
class Calculator:
    def __init__(self):
        self.result = 0
    # 주의: 이 부분의 코드를 __init__ 사용하여 result를 따로 정의하는 대신
    # 그냥
    # result = 0
    # 이렇게 정의하면 문제가 생김: 어떤 문제?

    def add(self, num):
        self.result += num
        return self.result

cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(3))
print(cal2.add(7))
````

어떤 문제?: 아래 예시를 실행시켜 보세요

````python
class Calculator:
    Cain = [] 
    
    def add(self, num):
        self.Cain.append(num)
        return sum(self.Cain)


cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(3))
print(cal2.add(7))
````

## 3. 상속 (기능 추가하기)

### 설명
* `FourCal` 클래스를 상속받아 제곱 기능(`pow`)을 추가한 클래스를 만드세요.
    * 상속 문법 이해

### 예시 출력

```sh
16
```

````python
class Calculator2:
    def __init__(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        return self.first + self.second

    def mul(self):
        return self.first * self.second


class powCal(Calculator2):
    def pow(self):
        return self.first ** self.second

calc = powCal(4, 2)
print(calc.pow())
````


## 4. 예외 처리

### 설명
* 나눗셈에서 0으로 나누면 에러가 발생합니다. `try / except`를 사용해 에러를 처리하세요.
    * `ZeroDivisionError` 예외 처리

### 예시 출력

```sh
0
```

````python
class SafeFourCal(Calculator2):
    def div(self):
        try:
            return self.first / self.second
        except ZeroDivisionError:
            return 0

calc = SafeFourCal(4, 0)
print(calc.div())
````
